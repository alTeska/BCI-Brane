# Data Visualization And Statistics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats, integrate

%matplotlib inline
```

## Load Data

```python
path = '../data/preprocessed/'
data = pd.read_csv(path+'CarloAll.csv', ',')
data.head()
```

# Plots

## Melted Dataset + Distribution for all labels

```python
melted = pd.melt(data, id_vars=['Time', 'id', 'label'], value_name="EEG")
melted.head()
```

```python
g = sns.FacetGrid(data=melted, size=8, col='label')
g.map(sns.distplot, 'EEG');
```

## Example plot for each label

```python
mask = (data.id == 0)
g = sns.FacetGrid(data=data[mask], size=6, aspect=2)
g.map(plt.plot, 'Time', 'AF3');

mask = (data.id == 1)
g = sns.FacetGrid(data=data[mask], size=6, aspect=2)
g.map(plt.plot, 'Time', 'AF3');

mask = (data.id == 3)
g = sns.FacetGrid(data=data[mask], size=6, aspect=2)
g.map(plt.plot, 'Time', 'AF3');
```
