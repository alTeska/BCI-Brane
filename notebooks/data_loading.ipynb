```python
import numpy as np
import pandas as pd
from glob import glob

%matplotlib inline
```

## Load Data

```python
# load and sort the paths
path = '../data/raw/'
sampling_rate = 128

fnames = glob(path+'*.csv') # geet paths and save them
fnames.sort()
len(fnames)
```

```python
fnames[:5]
```

### Save Into Data Frames 
Separation of the ',' and ';' cases.

```python
experiments = {}
for i in np.arange(0, 20):
    experiments[i] = pd.read_csv(fnames[i], ',')

for i in np.arange(20, len(fnames)):
    experiments[i] = pd.read_csv(fnames[i], ';') 
    
experiments[0].head()
```

```python
experiments[199].head()
```

## Data Frame Operations
### Label The Experiments Separately

```python
k = 0
for i in np.arange(0,160):
    experiments[i]['id'] = i
    experiments[i]['label'] = k
    k += 1
    if k==4:  k = 0

for i in np.arange(0,160):
    experiments[i]['label'].replace(3,-1, inplace=True)
    experiments[i]['label'].replace(2, 0, inplace=True)
        
k = 1
for i in np.arange(160, 200):
    experiments[i]['id'] = i
    experiments[i]['label'] = k
    if   k== 1: k=-1
    elif k==-1: k= 1
```

```python
# experiments[160].head()
experiments[161].head()
```

### Drop not needed columns

```python
for i in np.arange(0,200):
    experiments[i].drop(['Sampling Rate'], axis=1, inplace=True)
```

### Merging into one Dataset

```python
data = pd.DataFrame()

for i in np.arange(0,200):
    data = pd.concat([data, experiments[i]], axis=0, ignore_index=True)

data.rename(index=str, columns={"Time (s)": "Time"}, inplace=True)
data.head()
```

```python
len(data)
```

## Save The Data

```python
path2 = '../data/preprocessed/'
data.to_csv(path2+'CarloAll.csv', sep=',', index=False)
```
